```python
plt.figure(figsize=(12, 6))
education_counts = df['education'].value_counts()

# Создание bar plot
sns.barplot(x=education_counts.values, y=education_counts.index, palette='viridis')
plt.title('Распределение людей по уровню образования')
plt.xlabel('Количество людей')
plt.ylabel('Уровень образования')
plt.tight_layout()
plt.show()

# Вывод точных чисел
print("Количество людей по уровням образования:")
print(education_counts)
```

**Задание 10: А теперь давайте посмотрим на распределение возрастов. Сколько возрастных отрезков имеют больше 1600 значений? (Одним "отрезков" является один столбец при стандартных настройках seaborn/matplotlib)
P.S. Необязательное задание, хотя лучше сделать, если собираетесь сдавать до мягкого, я хоты бы посмотрю**

```python
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(df['age'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Распределение возрастов в датасете')
plt.xlabel('Возраст')
plt.ylabel('Количество людей')
plt.grid(True, alpha=0.3)

# Подсчет столбцов с более чем 1600 значений
count_above_1600 = np.sum(n > 1600)
print(f"Количество возрастных отрезков с более чем 1600 значений: {count_above_1600}")

# Визуальное выделение столбцов с >1600 значений
for i, patch in enumerate(patches):
    if n[i] > 1600:
        patch.set_facecolor('red')
    else:
        patch.set_facecolor('skyblue')

# Добавление горизонтальной линии на уровне 1600
plt.axhline(y=1600, color='red', linestyle='--', alpha=0.7, label='1600 человек')
plt.legend()

plt.tight_layout()
plt.show()

# Дополнительная информация
print(f"\nОбщее количество биннов: {len(n)}")
print(f"Максимальное количество в одном бинне: {n.max():.0f}")
print(f"Минимальное количество в одном бинне: {n.min():.0f}")
print(f"Среднее количество в бинне: {n.mean():.1f}")

# Покажем какие именно бинны имеют >1600 значений
print("\nБинны с более чем 1600 значений:")
for i, count in enumerate(n):
    if count > 1600:
        print(f"Бинн {i+1}: возраст {bins[i]:.1f}-{bins[i+1]:.1f}, количество: {count:.0f}")
```
